Que 4. Describe the architecture and functioning of a Recurrent Natural Network (RNN).

Ans - A Recurrent Neural Network (RNN) is a type of neural network designed to handle sequential data (like text, speech, or time-series). Unlike standard feedforward networks, RNNs have a “memory” that captures information from previous steps in the sequence.

---

🔹 1. Architecture of RNN

The core idea of an RNN is the recurrent connection, where the output from the previous time step is fed back into the network.

Basic Components:
- Input layer (xₜ): Takes input at time step t
- Hidden layer (hₜ): Stores information (memory)
- Output layer (yₜ): Produces output at time t

Key Feature:

The hidden state is updated using both:
- Current input (xₜ)
- Previous hidden state (hₜ₋₁)

Mathematical Representation:

  WhatsApp Image 2026-03-27 at 2.44.00 PM.jpeg

  ---

🔹 2. Unrolled Architecture

RNN can be visualized in two ways:

Folded View:
- Shows one unit with a loop (feedback connection)

Unrolled View:
- Expands across time steps:

    x1 → h1 → y1
     ↓
    x2 → h2 → y2
     ↓
    x3 → h3 → y3

Each time step shares the same weights, making RNN efficient for sequences.

---

🔹 3. Functioning of RNN

Step-by-step process:
1. Input sequence is fed one element at a time
2. Hidden state is updated using current input + previous memory
3. Output is generated at each time step (or only at final step depending on task)
4. Information flows through time via hidden states

---

🔹 4. Training of RNN

RNNs are trained using:

🔸 Backpropagation Through Time (BPTT)
- Extension of backpropagation
- Gradients are computed across all time steps
- Weights are updated based on cumulative error

---

🔹 5. Types of RNN Architectures

1. One-to-One
- Simple input → output (like basic NN)

2. One-to-Many
- Example: Image captioning

3. Many-to-One
- Example: Sentiment analysis

4. Many-to-Many
- Example: Machine translation

---

🔹 6. Advantages
- Handles sequential dependencies
- Useful for:
  - NLP (text, translation)
  - Speech recognition
  - Time-series prediction

---

🔹 7. Limitations
- Vanishing/Exploding gradient problem
- Difficulty learning long-term dependencies

👉 To solve this, advanced variants are used:
- LSTM (Long Short-Term Memory)
- GRU (Gated Recurrent Unit)

---

🔹 Final Summary

An RNN processes sequences by maintaining a hidden state that acts as memory, allowing it to capture patterns over time. It works step-by-step through sequences and is trained using Backpropagation Through Time, making it powerful for tasks involving temporal or sequential data.

Que 5. What is the key difference between LSTM and GRU networks in NLP applications ?

Ans - The key difference between LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) lies in their architecture and complexity, which affects performance and efficiency in NLP tasks.

---

🔹 1. Internal Structure

LSTM:
- Has 3 gates:
  - Input gate
  - Forget gate
  - Output gate
- Maintains two states:
  - Cell state (long-term memory)
  - Hidden state (short-term memory)

GRU:
- Has 2 gates:
  - Update gate
  - Reset gate
- Maintains only one state (combined cell + hidden state)

👉 Key difference:

GRU is a simplified version of LSTM.

---

🔹 2. Complexity & Speed
- LSTM → More complex, more parameters → slower to train
- GRU → Fewer parameters → faster and more efficient

👉 GRU is often preferred when you need speed and less computational cost.

---

🔹 3. Memory Handling
- LSTM → Better at capturing long-term dependencies due to separate cell state
- GRU → Good, but slightly less powerful for very long sequences

---

🔹 4. Performance in NLP
- LSTM:
  - Works well for tasks like machine translation, text generation
  - More control over memory
- GRU:
  - Performs similarly or sometimes better in many NLP tasks
  - Especially good when dataset is smaller or training time matters

---

🔹 5. Practical Insight
- If you want accuracy for long sequences → go with LSTM
- If you want faster training with similar performance → go with GRU

---

✅ Final Summary

| Feature        | LSTM          | GRU         |
| -------------- | ------------- | ----------- |
| Gates          | 3             | 2           |
| States         | Cell + Hidden | Hidden only |
| Complexity     | High          | Low         |
| Speed          | Slower        | Faster      |
| Long-term deps | Better        | Good        |

Que 8. Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).

# Dataset
texts = [
“I love this project”, #Positive
“This is an amazing experience”, #Positive
“I hate waiting in line”, #Negative
“This is the worst service”, #Negative
“Absolutely fantastic!” #Positive
]

labels = [1, 1, 0, 0, 1]

Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on this data. You may use Keras with TensorFlow backend.

Ans - Here’s a complete, simple implementation of an LSTM-based text classification model in Keras (TensorFlow backend) using your dummy dataset.

---

✅ Step-by-Step Implementation

1. Import Required Libraries


```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

---

2. Dataset

```python
texts = [
    "I love this project",          # Positive
    "This is an amazing experience",# Positive
    "I hate waiting in line",       # Negative
    "This is the worst service",    # Negative
    "Absolutely fantastic!"         # Positive
]

labels = [1, 1, 0, 0, 1]
```

---

3. Text Preprocessing (Tokenization + Padding)

```python
# Initialize tokenizer
tokenizer = Tokenizer(num_words=1000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)

# Convert text to sequences
sequences = tokenizer.texts_to_sequences(texts)

# Pad sequences
max_length = 10
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# Convert labels to numpy array
labels = np.array(labels)

print("Word Index:", tokenizer.word_index)
print("Padded Sequences:\n", padded_sequences)
```

---

4. Build the LSTM Model

```python
vocab_size = 1000
embedding_dim = 16

model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    LSTM(32),
    Dense(1, activation='sigmoid')  # Binary classification
])
```

---

5. Compile the Model

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

---

6. Train the Model


```python
model.fit(padded_sequences, labels, epochs=10, verbose=1)
```

---

7. Test the Model

```python
# Test sentences
test_texts = ["I really love this", "This is horrible"]

# Preprocess test data
test_seq = tokenizer.texts_to_sequences(test_texts)
test_pad = pad_sequences(test_seq, maxlen=max_length, padding='post')

# Predict
predictions = model.predict(test_pad)

for text, pred in zip(test_texts, predictions):
    print(f"{text} -> {'Positive' if pred > 0.5 else 'Negative'} ({pred[0]:.4f})")
```

---

🔍 Explanation (Short & Practical)
- Tokenizer → converts words into integers
- Padding → makes all sequences same length
- Embedding layer → converts words into dense vectors
- LSTM → captures sequence dependencies
- Sigmoid output → gives probability for binary classification

---

⚠️ Note
- Dataset is very small → model won’t generalize well (just for learning)
- For real tasks → use larger datasets + validation split

Que 9. Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:

“Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.”

Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.

Ans - Here’s a clean Python example using spaCy to build a simple NLP pipeline that performs tokenization, lemmatization, and named entity recognition (NER) on your paragraph.

---

✅ Step 1: Install spaCy and download model

    pip install spacy
    python -m spacy download en_core_web_sm

✅ Step 2: Python Program



```python
import spacy

# Load English NLP model
nlp = spacy.load("en_core_web_sm")

# Input text
text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of India’s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""

# Process the text
doc = nlp(text)

# ----------- TOKENIZATION & LEMMATIZATION -----------
print("TOKENS AND LEMMAS:\n")
for token in doc:
    print(f"Token: {token.text:15} Lemma: {token.lemma_}")

# ----------- NAMED ENTITY RECOGNITION -----------
print("\nNAMED ENTITIES:\n")
for ent in doc.ents:
    print(f"Entity: {ent.text:50} Label: {ent.label_}")
```

---

🔹 Tokenization + Lemmatization
- Splits text into words (tokens)
- Converts each word to its base form (lemma)

🔹 Named Entity Recognition (NER)

Detects real-world entities like:
- PERSON → Homi Jehangir Bhaba
- ORG → Tata Institute of Fundamental Research
- GPE → India

---

💡 Optional Improvement

If you want cleaner output (skip punctuation & stopwords):


```python
for token in doc:
    if not token.is_stop and not token.is_punct:
        print(f"{token.text:15} -> {token.lemma_}")
```
